```sql
-- Original NDS Store
USE NDS;
SELECT store.store_key, store.store_number, store.store_name, store_type.description AS store_type, address.address1, address.address2, address.address3, address.address4, city.city_name, state.state_name, address.post_code, 
                  country.country_name, phone_number.phone_number, store.web_site, region.region_name, division.division_name, store.source_system_code, store.create_timestamp, store.update_timestamp
FROM     store INNER JOIN
                  store_type ON store.store_type_key = store_type.store_type_key INNER JOIN
                  address ON store.address_key = address.address_key INNER JOIN
                  city ON address.city_key = city.city_key LEFT JOIN
                  state ON address.state_key = state.state_key INNER JOIN
                  country ON address.country_key = country.country_key INNER JOIN
                  phone_number ON store.phone_number_key = phone_number.phone_number_key LEFT JOIN
                  region ON store.region_key = region.region_key INNER JOIN
                  division ON region.division_key = division.division_key
-- WHERE  (store.create_timestamp >= ? AND store.create_timestamp < ? ) OR
 --                 (store.update_timestamp >= ? AND store.update_timestamp < ?)
```

```sql
-- Trouble shooting and fixed solution

USE NDS;
SELECT store.store_key, store.store_number, store.store_name, store_type.description AS store_type, address.address1, address.address2, address.address3, address.address4, city.city_name, state.state_name, address.post_code, 
                  country.country_name, phone_number.phone_number, store.web_site, region.region_name, division.division_name, store.source_system_code, store.create_timestamp, store.update_timestamp
FROM     store INNER JOIN store_type ON store.store_type_key = store_type.store_type_key                   
               INNER JOIN address ON store.address_key = address.address_key 
               LEFT JOIN city ON address.city_key = city.city_key 
               LEFT JOIN state ON address.state_key = state.state_key 
               LEFT JOIN country ON address.country_key = country.country_key
               INNER JOIN phone_number ON store.phone_number_key = phone_number.phone_number_key
               LEFT JOIN region ON store.region_key = region.region_key 
               INNER JOIN division ON region.division_key = division.division_key
WHERE  (store.create_timestamp >= ? AND store.create_timestamp < ? ) 
  OR (store.update_timestamp >= ? AND store.update_timestamp < ?)
```
